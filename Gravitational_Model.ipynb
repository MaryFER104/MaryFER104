The gravity model relate the amount of commerce with the size of countries, its
distance an its international trade.

Using the datasets:
- Locations
- Importations
- Exportations
- Gross domestic products
- Country names

The following work need to establish empirical evidence in Colombia, could
test any model to support or not the hypothesis, for instance:
$$ y_{ij} = β_0 + β_1 gpd_j + β_2 dist_{ij} $$
Where the $ y_{ij} $ is the amount of international trade (importations+exportations)
between the $ i $ country with the $ j $ country, $ gdp_j $ is the domestic product $ j $ and
$ dist_{ij} $ could be a measure of distance among the countries $ i $ and $ j $.

# Funciones con REGEX

```python
# Cambiar el nombre de las columnas que contengan Enero - Diciembre [Año]

def new_cnames_year(DataFrame):
  patron = r'[\w]+ . [\w]+'

  for nombre in DataFrame.columns:
    if re.match(patron, nombre):
      l10n = re.search(r'\d+', nombre).span()
      nuevo_nombre = nombre[l10n[0] : l10n[1]]
      DataFrame.rename(columns={nombre: nuevo_nombre}, inplace=True)
```

```python
new_cnames_year(df_importations)
df_importations.head()
```

```python
df_importations
```

```python
# Agregar el año anterior al nombre de las columnas tituladas Unnamed

def add_year(DataFrame):
  patron_un = r'Unnamed: \d+'
  patron_year = r'\d+'

  _prev_year = 0
  _nombres_columnas = DataFrame.columns.tolist()

  for nombre in _nombres_columnas:
    if re.fullmatch(patron_year, nombre):
      _prev_year = nombre

    if re.match(patron_un, nombre):
      _nuevo_nombre = _prev_year
      DataFrame.rename(columns={nombre: _nuevo_nombre}, inplace=True)
```

```python
add_year(df_importations)
df_importations.head()
```

```python
add_year(df_exportations)
df_exportations.head()
```
